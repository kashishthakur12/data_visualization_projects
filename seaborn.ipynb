```python
#pairgrid ->subplot grid for plottingoairwise relationships in a dataset

graph = sns.PairGrid(data = penguins, hue = "sex", palette = "Set2")
graph.map_upper(sns.scatterplot)#uppertriangle
graph.map_lower(sns.kdeplot)#lowertriangle
graph.map_diag(sns.histplot)#diagnol
graph.add_legend()
```
