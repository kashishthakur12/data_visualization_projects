10. GROUPBY AND AGGREGATIONS

```python
import matplotlib.pyplot as plt
data = {
    "region": ["East","East","West","West","South","South"],
    "product": ["A","B","A","B","A","B"],
    "sales": [100,150,200,250,300,350]
}
df = pd.DataFrame(data)
print(df)

# Group by single column
print(df.groupby("region").mean(numeric_only=True))


# Group by multiple columns
print(df.groupby(["region","product"]).sum())

# Value counts
print(df["product"].value_counts())
```

11. PIVOT TABLES AND CROSSTAB

```python
# Pivot Table
print(df.pivot_table(values="sales", index="region", columns="product", aggfunc="sum"))

# Crosstab
print(pd.crosstab(df["region"], df["product"]))
```

12. CONCAT AND MERGE

```python
df1 = pd.DataFrame({"key":[1,2,3], "val1":["A","B","C"]})
df2 = pd.DataFrame({"key":[2,3,4], "val2":["X","Y","Z"]})

# Row-wise concat
print(pd.concat([df1,df2]))

# Column-wise concat
print(pd.concat([df1,df2], axis=1))

# Inner merge
print(pd.merge(df1, df2, on="key"))

# Left merge
print(pd.merge(df1, df2, how="left"))

# Right merge
print(pd.merge(df1, df2, how="right"))

# Outer merge
print(pd.merge(df1, df2, how="outer"))
```

13.RESHAPING

```python
df3 = pd.DataFrame({
    "id":[1,1,2,2],
    "feature":["height","weight","height","weight"],
    "value":[5.5,60,6.0,70]
})

# Melt (unpivot)
print(df3.melt())

# Pivot
print(df3.pivot(index="id", columns="feature", values="value"))

# Stack / Unstack
stacked = df3.pivot(index="id", columns="feature", values="value").stack()
print(stacked)
print(stacked.unstack())
```

14. DATATIME HANDLING

```python
df4 = pd.DataFrame({
    "date":["2023-01-01","2023-02-15","2023-03-20"]
})

df4["date"] = pd.to_datetime(df4["date"])
df4["year"] = df4["date"].dt.year
df4["month"] = df4["date"].dt.month
df4["day"] = df4["date"].dt.day
df4["weekday"] = df4["date"].dt.day_name()
print(df4)
```

15.STRING OPERATION

```python
df6 = pd.DataFrame({"col":[10,20,30,40,50], "col2":[5,15,25,35,45]})

df6["col"].plot(kind="line", title="Line Plot")
plt.show()

df6["col"].plot(kind="bar", title="Bar Plot")
plt.show()

df6["col"].plot(kind="hist", title="Histogram")
plt.show()

df6["col"].plot(kind="box", title="Box Plot")
plt.show()

df6.plot.scatter(x="col", y="col2", title="Scatter Plot")
plt.show()

df6["col"].value_counts().plot(kind="pie", title="Pie Chart")
plt.show()
```
