```python
#plt.plot(x,y,color = 'color name',linestyle = 'line_style',linewidth = value, marker = 'marker symbol' , label = 'label name)
#plt.xlabel('text')
#plt,xlim(start,end)

months = [1,2,3,4]
sales = [1000,2000,1200,1800]

plt.plot(months,sales,color = 'red', linestyle = '--', linewidth = 2, marker = 'o',label = '2025 sales data')
plt.xlabel('months')
plt.ylabel('sales per month')
plt.title('monthly sales data report')
plt.legend(loc = 'upper left', fontsize = 12)
plt.grid(color = 'green', linestyle = ':', linewidth = 0.9)
plt.xlim(1,4)
plt.ylim(0,2000)
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])
plt.show()
```

```python
#plt.bar(x,height,'color name',width = value, label = 'label name')
import matplotlib.pyplot as plt

product = ['A','B','C','D','E']
sales = [1000,1500,800,1200,1677]

plt.bar(product, sales,color = 'blue',label = 'sales 2026')
plt.xlabel('product')
plt.ylabel('sales')
plt.title('product sale comparison')
plt.legend()
plt.show()
```

```python
#for horiszontal representation
#plt.barh (h is add)

import matplotlib.pyplot as plt

product = ['A','B','C','D','E']
sales = [1000,1500,800,1200,1677]

plt.barh(product, sales,color = 'blue',label = 'sales 2026')
plt.xlabel('product')
plt.ylabel('sales')
plt.title('product sale comparison')
plt.legend()
plt.show()
```

```python
#plt.histogram(data, bins = 'numbers_of_bins, color='colorname',edgecolor='black')
import matplotlib.pyplot as plt

scores = [45,67,89,34,56,78,1,4,90,87,65,32,79,34,21,11,6,88,99,88,99,88,99,88,99]

plt.hist(scores,bins = 8,color='purple',edgecolor = 'black')
plt.xlabel('score Range')
plt.ylabel('Number of students')
plt.title('Score distribution of student')
plt.show()
```

```python
# plt.scatter(x,y,color = 'color_name',marker= 'marker style',label = 'label name')
import matplotlib.pyplot as plt

hours_studies = [1,2,3,4,5,6,7,8]
exam_scores = [50,56,78,90,92,93,94,97]

#scatter plot
plt.scatter(hours_studies,exam_scores,color='green', marker= '^', label = 'student data')
plt.xlabel('hours studied')
plt.ylabel('Exam scores')
plt.title('relationship between study time and exam score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# plt.scatter(x,y,color = 'color_name',marker= 'marker style',label = 'label name')
import matplotlib.pyplot as plt

#scatter plot
plt.scatter([1,2,3,4],[34,55,78,67],color='blue', marker= '^', label = 'class A')
plt.scatter([1,2,3,4],[44,65,98,17],color='orange', marker= '^', label = 'class B')

plt.xlabel('hours studied')
plt.ylabel('Exam scores')
plt.title('comparison between two classes')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#plt.subplot(nrows,ncols,index)

import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,30,40]

plt.subplot(1,2,1)#1 row,2column,1st subplot
plt.plot(x,y)
plt.title('Line chart')

plt.subplot(1,2,2)#1 row,2column,2nd subplot
plt.bar(x,y)
plt.title('BarChart')

#plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,30,40]
 
# fig, ax = plt.subplots(nrows,ncols,figsize = (width,height))
fig, ax = plt.subplots(1,2,figsize = (10,5)) 

x = [1,2,3,4]
y = [10,20,30,34]

ax[0].plot(x,y)
ax[0].set_title('Line plot')

ax[1].bar(x,y)
ax[1].set_title('Bar chart')

fig.suptitle('comparison between line and bar charts')
plt.tight_layout()
plt.show()
```
